```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
%matplotlib inline
```

```python
DA=pd.read_csv('Dataset A.csv')
```

```python
DA
```

```python
DA.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=DA[['Total Purchase from App (Rupees/year)','Total Purchase from Website (Rupees/year)','Average Time on App (Minutes/Day)','Average Time on Website (Minutes/Day)','Length of Membership (Year)','Yearly Amount Spent (Minutes/year)','Total Purchase of each customer (Rupees/year)']]
y=DA['Ratings given by customer (Out of 5)']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=25)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```
